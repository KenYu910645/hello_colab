```python
import cv2
import os
from google.colab.patches import cv2_imshow

# get file names
input_path = "/content/gdrive/My Drive/DeRaindrop/homemade_image/"
output_path = "/content/gdrive/My Drive/DeRaindrop/homemade_image_output/"
input_list = sorted(os.listdir(input_path))

for i in range(len(input_list)):
    print ('Processing image: %s'%(input_list[i]))
    image = cv2.imread(input_path + input_list[i])
    # cv2_imshow(image)
    image = cv2.resize(image, (720, 480), interpolation=cv2.INTER_AREA)
    cv2.imwrite(output_path + input_list[i], image)
```
